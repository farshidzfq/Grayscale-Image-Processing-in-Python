```python
# Import required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload the image
print("Please upload a grayscale image:")
uploaded = files.upload()

# Get the uploaded file name
image_path = list(uploaded.keys())[0]

# Step 2: Load the grayscale image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.figure(figsize=(10, 5))
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()

# Step 3: Binarize the image
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)

# Display the binary image
plt.figure(figsize=(10, 5))
plt.title("Binarized Image")
plt.imshow(binary_image, cmap='gray')
plt.axis('off')
plt.show()

# Step 4: Apply Morphological Operations to Remove Noise
kernel = np.ones((3, 3), np.uint8)
cleaned_image = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel, iterations=1)

# Display the cleaned image
plt.figure(figsize=(10, 5))
plt.title("Cleaned Image (Noise Removed)")
plt.imshow(cleaned_image, cmap='gray')
plt.axis('off')
plt.show()
```
